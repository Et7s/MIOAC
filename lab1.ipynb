```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers, Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import tensorflow as tf
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

# â­â­ Ğ”ĞĞ‘ĞĞ’Ğ›Ğ¯Ğ•Ğœ ĞŸĞ ĞĞ’Ğ•Ğ ĞšĞ£ Ğ˜ ĞĞĞ¡Ğ¢Ğ ĞĞ™ĞšĞ£ GPU â­â­
print("ĞŸÑ€Ğ¾Ğ²ĞµÑ€ĞºĞ° Ğ´Ğ¾ÑÑ‚ÑƒĞ¿Ğ½Ğ¾ÑÑ‚Ğ¸ GPU...")
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        # Ğ Ğ°Ğ·Ñ€ĞµÑˆĞ°ĞµĞ¼ Ñ€Ğ¾ÑÑ‚ Ğ¿Ğ°Ğ¼ÑÑ‚Ğ¸ GPU
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"ĞĞ°Ğ¹Ğ´ĞµĞ½Ğ¾ GPU: {len(gpus)} ÑƒÑÑ‚Ñ€Ğ¾Ğ¹ÑÑ‚Ğ²")
        print("Ğ˜ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµÑ‚ÑÑ GPU Ğ´Ğ»Ñ ÑƒÑĞºĞ¾Ñ€ĞµĞ½Ğ¸Ñ Ğ²Ñ‹Ñ‡Ğ¸ÑĞ»ĞµĞ½Ğ¸Ğ¹! ğŸš€")
    except RuntimeError as e:
        print(f"ĞÑˆĞ¸Ğ±ĞºĞ° Ğ½Ğ°ÑÑ‚Ñ€Ğ¾Ğ¹ĞºĞ¸ GPU: {e}")
else:
    print("GPU Ğ½Ğµ Ğ½Ğ°Ğ¹Ğ´ĞµĞ½Ğ¾, Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµÑ‚ÑÑ CPU")

# Ğ£ÑÑ‚Ğ°Ğ½Ğ°Ğ²Ğ»Ğ¸Ğ²Ğ°ĞµĞ¼ Ñ‚Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚ÑŒ mixed precision Ğ´Ğ»Ñ Ğ±Ğ¾Ğ»ÑŒÑˆĞµĞ¹ ÑÑ„Ñ„ĞµĞºÑ‚Ğ¸Ğ²Ğ½Ğ¾ÑÑ‚Ğ¸ Ğ½Ğ° GPU
# â­ Ğ˜Ğ¡ĞŸĞ ĞĞ’Ğ›Ğ•ĞĞ˜Ğ•: Ğ”Ğ¾Ğ±Ğ°Ğ²Ğ»ÑĞµĞ¼ Ğ¿Ñ€Ğ¾Ğ²ĞµÑ€ĞºÑƒ Ğ¿ĞµÑ€ĞµĞ´ ÑƒÑÑ‚Ğ°Ğ½Ğ¾Ğ²ĞºĞ¾Ğ¹ mixed precision
try:
    policy = tf.keras.mixed_precision.Policy('mixed_float16')
    tf.keras.mixed_precision.set_global_policy(policy)
    print(f"Ğ£ÑÑ‚Ğ°Ğ½Ğ¾Ğ²Ğ»ĞµĞ½Ğ° Ğ¿Ğ¾Ğ»Ğ¸Ñ‚Ğ¸ĞºĞ° Ğ²Ñ‹Ñ‡Ğ¸ÑĞ»ĞµĞ½Ğ¸Ğ¹: {policy.name}")
except Exception as e:
    print(f"Mixed precision Ğ½Ğµ Ğ´Ğ¾ÑÑ‚ÑƒĞ¿ĞµĞ½: {e}")
    print("Ğ˜ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµÑ‚ÑÑ ÑÑ‚Ğ°Ğ½Ğ´Ğ°Ñ€Ñ‚Ğ½Ğ°Ñ Ñ‚Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚ÑŒ float32")

# Ğ¢ĞµĞ¿ĞµÑ€ÑŒ ÑƒÑÑ‚Ğ°Ğ½Ğ°Ğ²Ğ»Ğ¸Ğ²Ğ°ĞµĞ¼ skimage (ĞµÑĞ»Ğ¸ Ğ½Ğµ ÑƒÑÑ‚Ğ°Ğ½Ğ¾Ğ²Ğ»ĞµĞ½)
try:
    from skimage.metrics import peak_signal_noise_ratio as psnr
    from skimage.metrics import structural_similarity as ssim
except ImportError:
    print("Ğ£ÑÑ‚Ğ°Ğ½Ğ¾Ğ²ĞºĞ° scikit-image...")
    import subprocess
    import sys
    subprocess.check_call([sys.executable, "-m", "pip", "install", "scikit-image"])
    from skimage.metrics import peak_signal_noise_ratio as psnr
    from skimage.metrics import structural_similarity as ssim

# 1. Ğ£Ğ»ÑƒÑ‡ÑˆĞµĞ½Ğ½Ğ°Ñ Ğ·Ğ°Ğ³Ñ€ÑƒĞ·ĞºĞ° Ğ¸ Ğ¿Ğ¾Ğ´Ğ³Ğ¾Ñ‚Ğ¾Ğ²ĞºĞ° Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…
print("\nĞ—Ğ°Ğ³Ñ€ÑƒĞ·ĞºĞ° Ğ¸ Ğ¿Ğ¾Ğ´Ğ³Ğ¾Ñ‚Ğ¾Ğ²ĞºĞ° Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…...")
(x_train, _), (x_test, _) = keras.datasets.cifar10.load_data()

# ĞĞ¾Ñ€Ğ¼Ğ°Ğ»Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ Ğ¸ Ğ¿Ñ€ĞµĞ¾Ğ±Ñ€Ğ°Ğ·Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ Ğ² float32
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Ğ”Ğ¾Ğ±Ğ°Ğ²Ğ»ÑĞµĞ¼ Ğ²Ğ°Ğ»Ğ¸Ğ´Ğ°Ñ†Ğ¸Ğ¾Ğ½Ğ½ÑƒÑ Ğ²Ñ‹Ğ±Ğ¾Ñ€ĞºÑƒ
x_val = x_test[:5000]
x_test = x_test[5000:]

print(f"Ğ Ğ°Ğ·Ğ¼ĞµÑ€Ñ‹ Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…: Train: {x_train.shape}, Val: {x_val.shape}, Test: {x_test.shape}")

# 2. Ğ£Ğ»ÑƒÑ‡ÑˆĞµĞ½Ğ½Ğ°Ñ Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ñ Ğ´Ğ¾Ğ±Ğ°Ğ²Ğ»ĞµĞ½Ğ¸Ñ ÑˆÑƒĞ¼Ğ° Ñ Ñ€Ğ°Ğ·Ğ½Ñ‹Ğ¼Ğ¸ Ñ‚Ğ¸Ğ¿Ğ°Ğ¼Ğ¸ ÑˆÑƒĞ¼Ğ°
def add_mixed_noise(images, gaussian_noise_factor=0.3, salt_pepper_ratio=0.02):
    """
    Ğ”Ğ¾Ğ±Ğ°Ğ²Ğ»ÑĞµÑ‚ ÑĞ¼ĞµÑˆĞ°Ğ½Ğ½Ñ‹Ğ¹ ÑˆÑƒĞ¼: Ğ³Ğ°ÑƒÑÑĞ¾Ğ²ÑĞºĞ¸Ğ¹ + ÑĞ¾Ğ»ÑŒ-Ğ¿ĞµÑ€ĞµÑ†
    """
    noisy = images.copy()

    # Ğ“Ğ°ÑƒÑÑĞ¾Ğ²ÑĞºĞ¸Ğ¹ ÑˆÑƒĞ¼
    gaussian_noise = np.random.normal(loc=0.0, scale=1.0, size=images.shape)
    noisy += gaussian_noise_factor * gaussian_noise

    # Ğ¨ÑƒĞ¼ "ÑĞ¾Ğ»ÑŒ-Ğ¿ĞµÑ€ĞµÑ†"
    salt_pepper = np.random.random(images.shape[:3])
    noisy[salt_pepper < salt_pepper_ratio/2] = 0.0  # ĞŸĞµÑ€ĞµÑ†
    noisy[salt_pepper > 1 - salt_pepper_ratio/2] = 1.0  # Ğ¡Ğ¾Ğ»ÑŒ

    return np.clip(noisy, 0.0, 1.0)

# Ğ¡Ğ¾Ğ·Ğ´Ğ°ĞµĞ¼ Ğ·Ğ°ÑˆÑƒĞ¼Ğ»ĞµĞ½Ğ½Ñ‹Ğµ Ğ²ĞµÑ€ÑĞ¸Ğ¸
print("Ğ”Ğ¾Ğ±Ğ°Ğ²Ğ»ĞµĞ½Ğ¸Ğµ ÑˆÑƒĞ¼Ğ° Ğº Ğ´Ğ°Ğ½Ğ½Ñ‹Ğ¼...")
x_train_noisy = add_mixed_noise(x_train)
x_val_noisy = add_mixed_noise(x_val)
x_test_noisy = add_mixed_noise(x_test)

# 3. Ğ£Ğ»ÑƒÑ‡ÑˆĞµĞ½Ğ½Ğ°Ñ Ğ°Ñ€Ñ…Ğ¸Ñ‚ĞµĞºÑ‚ÑƒÑ€Ğ° Ğ°Ğ²Ñ‚Ğ¾ÑĞ½ĞºĞ¾Ğ´ĞµÑ€Ğ° Ñ skip-connections (Ğ¿Ğ¾Ñ…Ğ¾Ğ¶Ğµ Ğ½Ğ° U-Net)
def build_denoising_autoencoder(input_shape=(32, 32, 3)):
    """
    Ğ¡Ñ‚Ñ€Ğ¾Ğ¸Ñ‚ ÑƒĞ»ÑƒÑ‡ÑˆĞµĞ½Ğ½Ñ‹Ğ¹ Ğ°Ğ²Ñ‚Ğ¾ÑĞ½ĞºĞ¾Ğ´ĞµÑ€ Ğ´Ğ»Ñ ÑˆÑƒĞ¼Ğ¾Ğ¿Ğ¾Ğ´Ğ°Ğ²Ğ»ĞµĞ½Ğ¸Ñ
    """
    inputs = layers.Input(shape=input_shape)

    # Ğ­Ğ½ĞºĞ¾Ğ´ĞµÑ€
    x1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    x1 = layers.BatchNormalization()(x1)
    x1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x1)
    x1 = layers.BatchNormalization()(x1)
    p1 = layers.MaxPooling2D((2, 2))(x1)

    x2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(p1)
    x2 = layers.BatchNormalization()(x2)
    x2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x2)
    x2 = layers.BatchNormalization()(x2)
    p2 = layers.MaxPooling2D((2, 2))(x2)

    x3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(p2)
    x3 = layers.BatchNormalization()(x3)
    x3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(x3)
    x3 = layers.BatchNormalization()(x3)

    # Ğ”ĞµĞºĞ¾Ğ´ĞµÑ€ Ñ skip-connections
    u2 = layers.UpSampling2D((2, 2))(x3)
    c2 = layers.concatenate([u2, x2])
    x4 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c2)
    x4 = layers.BatchNormalization()(x4)
    x4 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x4)
    x4 = layers.BatchNormalization()(x4)

    u1 = layers.UpSampling2D((2, 2))(x4)
    c1 = layers.concatenate([u1, x1])
    x5 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c1)
    x5 = layers.BatchNormalization()(x5)
    x5 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x5)
    x5 = layers.BatchNormalization()(x5)

    # â­ Ğ˜Ğ¡ĞŸĞ ĞĞ’Ğ›Ğ•ĞĞ˜Ğ•: Ğ˜ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµĞ¼ sigmoid Ğ²Ğ¼ĞµÑÑ‚Ğ¾ linear Ğ´Ğ»Ñ Ğ¿Ñ€Ğ°Ğ²Ğ¸Ğ»ÑŒĞ½Ğ¾Ğ³Ğ¾ Ğ´Ğ¸Ğ°Ğ¿Ğ°Ğ·Ğ¾Ğ½Ğ°
    outputs = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same', dtype='float32')(x5)

    model = Model(inputs, outputs)
    return model

# Ğ¡Ğ¾Ğ·Ğ´Ğ°ĞµĞ¼ Ğ¼Ğ¾Ğ´ĞµĞ»ÑŒ Ñ Ğ¾Ğ¿Ñ‚Ğ¸Ğ¼Ğ¸Ğ·Ğ°Ñ†Ğ¸ĞµĞ¹ Ğ´Ğ»Ñ GPU
print("Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ½Ğ¸Ğµ ÑƒĞ»ÑƒÑ‡ÑˆĞµĞ½Ğ½Ğ¾Ğ³Ğ¾ Ğ°Ğ²Ñ‚Ğ¾ÑĞ½ĞºĞ¾Ğ´ĞµÑ€Ğ°...")
autoencoder = build_denoising_autoencoder()

# ĞšĞ¾Ğ¼Ğ¿Ğ¸Ğ»Ğ¸Ñ€ÑƒĞµĞ¼ Ñ Ğ¾Ğ¿Ñ‚Ğ¸Ğ¼Ğ¸Ğ·Ğ°Ñ†Ğ¸ĞµĞ¹ Ğ´Ğ»Ñ GPU
autoencoder.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae']
)

print(autoencoder.summary())

# 4. Ğ£Ğ»ÑƒÑ‡ÑˆĞµĞ½Ğ½Ñ‹Ğµ callback'Ñ‹ Ğ´Ğ»Ñ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ñ
callbacks = [
    EarlyStopping(patience=10, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(factor=0.5, patience=5, min_lr=1e-7, verbose=1),
    keras.callbacks.ModelCheckpoint('best_denoising_autoencoder.h5',
                                   save_best_only=True,
                                   save_weights_only=False,
                                   monitor='val_loss')
]

# 5. ĞĞ±ÑƒÑ‡ĞµĞ½Ğ¸Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸ Ñ Ğ¾Ğ¿Ñ‚Ğ¸Ğ¼Ğ¸Ğ·Ğ°Ñ†Ğ¸ĞµĞ¹ Ğ´Ğ»Ñ GPU
print("ĞĞ°Ñ‡Ğ°Ğ»Ğ¾ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ñ Ğ½Ğ° GPU...")

# Ğ˜ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµĞ¼ tf.data Ğ´Ğ»Ñ Ğ±Ğ¾Ğ»ĞµĞµ ÑÑ„Ñ„ĞµĞºÑ‚Ğ¸Ğ²Ğ½Ğ¾Ğ¹ Ğ·Ğ°Ğ³Ñ€ÑƒĞ·ĞºĞ¸ Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ… Ğ½Ğ° GPU
train_dataset = tf.data.Dataset.from_tensor_slices((x_train_noisy, x_train))
train_dataset = train_dataset.shuffle(1000).batch(64).prefetch(tf.data.AUTOTUNE)

val_dataset = tf.data.Dataset.from_tensor_slices((x_val_noisy, x_val))
val_dataset = val_dataset.batch(64).prefetch(tf.data.AUTOTUNE)

history = autoencoder.fit(
    train_dataset,
    epochs=30,  # ĞœĞ¾Ğ¶Ğ½Ğ¾ ÑƒĞ²ĞµĞ»Ğ¸Ñ‡Ğ¸Ñ‚ÑŒ, Ñ‚Ğ°Ğº ĞºĞ°Ğº Ğ½Ğ° GPU Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ğµ Ğ±Ñ‹ÑÑ‚Ñ€ĞµĞµ
    validation_data=val_dataset,
    callbacks=callbacks,
    verbose=1
)

# 6. Ğ¤ÑƒĞ½ĞºÑ†Ğ¸Ğ¸ Ğ´Ğ»Ñ Ğ¾Ñ†ĞµĞ½ĞºĞ¸ ĞºĞ°Ñ‡ĞµÑÑ‚Ğ²Ğ°
def calculate_metrics(original, denoised):
    """Ğ’Ñ‹Ñ‡Ğ¸ÑĞ»ÑĞµÑ‚ Ğ¼ĞµÑ‚Ñ€Ğ¸ĞºĞ¸ ĞºĞ°Ñ‡ĞµÑÑ‚Ğ²Ğ° PSNR Ğ¸ SSIM"""
    psnr_values = []
    ssim_values = []

    for i in range(len(original)):
        # PSNR (Peak Signal-to-Noise Ratio)
        psnr_val = psnr(original[i], denoised[i], data_range=1.0)
        psnr_values.append(psnr_val)

        # SSIM (Structural Similarity Index)
        ssim_val = ssim(original[i], denoised[i],
                       data_range=1.0, channel_axis=-1, win_size=3)
        ssim_values.append(ssim_val)

    return np.mean(psnr_values), np.mean(ssim_values)

def evaluate_model(model, x_clean, x_noisy):
    """ĞŸĞ¾Ğ»Ğ½Ğ°Ñ Ğ¾Ñ†ĞµĞ½ĞºĞ° Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸"""
    # ĞŸÑ€ĞµĞ´ÑĞºĞ°Ğ·Ğ°Ğ½Ğ¸Ğµ Ğ½Ğ° GPU
    denoised = model.predict(x_noisy, verbose=0, batch_size=64)

    # â­ Ğ˜Ğ¡ĞŸĞ ĞĞ’Ğ›Ğ•ĞĞ˜Ğ•: ĞŸÑ€ĞµĞ¾Ğ±Ñ€Ğ°Ğ·ÑƒĞµĞ¼ Ğ² float32 Ğ´Ğ»Ñ ÑĞ¾Ğ²Ğ¼ĞµÑÑ‚Ğ¸Ğ¼Ğ¾ÑÑ‚Ğ¸
    denoised = denoised.astype('float32')

    # Ğ’Ñ‹Ñ‡Ğ¸ÑĞ»ĞµĞ½Ğ¸Ğµ Ğ¼ĞµÑ‚Ñ€Ğ¸Ğº
    mse = np.mean((x_clean - denoised) ** 2)
    psnr_score, ssim_score = calculate_metrics(x_clean, denoised)

    print(f"\n=== Ğ Ğ•Ğ—Ğ£Ğ›Ğ¬Ğ¢ĞĞ¢Ğ« ĞĞ¦Ğ•ĞĞšĞ˜ ===")
    print(f"MSE: {mse:.6f}")
    print(f"PSNR: {psnr_score:.2f} dB")
    print(f"SSIM: {ssim_score:.4f}")

    return denoised, mse, psnr_score, ssim_score

# 7. Ğ’Ğ¸Ğ·ÑƒĞ°Ğ»Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ Ñ€ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚Ğ¾Ğ²
def plot_comparison(original, noisy, denoised, metrics, num_samples=5):
    """Ğ£Ğ»ÑƒÑ‡ÑˆĞµĞ½Ğ½Ğ°Ñ Ğ²Ğ¸Ğ·ÑƒĞ°Ğ»Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ ÑÑ€Ğ°Ğ²Ğ½ĞµĞ½Ğ¸Ñ"""
    plt.figure(figsize=(15, 10))

    for i in range(num_samples):
        # ĞÑ€Ğ¸Ğ³Ğ¸Ğ½Ğ°Ğ»ÑŒĞ½Ğ¾Ğµ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ğµ
        plt.subplot(4, num_samples, i + 1)
        # â­ Ğ˜Ğ¡ĞŸĞ ĞĞ’Ğ›Ğ•ĞĞ˜Ğ•: Ğ£Ğ±ĞµĞ¶Ğ´Ğ°ĞµĞ¼ÑÑ, Ñ‡Ñ‚Ğ¾ Ğ´Ğ°Ğ½Ğ½Ñ‹Ğµ Ğ² Ğ¿Ñ€Ğ°Ğ²Ğ¸Ğ»ÑŒĞ½Ğ¾Ğ¼ Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ‚Ğµ
        plt.imshow(np.clip(original[i], 0, 1))
        plt.title("Original")
        plt.axis('off')

        # Ğ—Ğ°ÑˆÑƒĞ¼Ğ»ĞµĞ½Ğ½Ğ¾Ğµ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ğµ
        plt.subplot(4, num_samples, i + 1 + num_samples)
        plt.imshow(np.clip(noisy[i], 0, 1))
        plt.title("Noisy")
        plt.axis('off')

        # Ğ’Ğ¾ÑÑÑ‚Ğ°Ğ½Ğ¾Ğ²Ğ»ĞµĞ½Ğ½Ğ¾Ğµ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ğµ
        plt.subplot(4, num_samples, i + 1 + 2 * num_samples)
        # â­ Ğ˜Ğ¡ĞŸĞ ĞĞ’Ğ›Ğ•ĞĞ˜Ğ•: ĞŸÑ€ĞµĞ¾Ğ±Ñ€Ğ°Ğ·ÑƒĞµĞ¼ Ğ¸ Ğ¾Ğ³Ñ€Ğ°Ğ½Ğ¸Ñ‡Ğ¸Ğ²Ğ°ĞµĞ¼ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ¸Ñ
        denoised_display = np.clip(denoised[i], 0, 1)
        plt.imshow(denoised_display)
        plt.title("Denoised")
        plt.axis('off')

        # Ğ Ğ°Ğ·Ğ½Ğ¸Ñ†Ğ° (Ğ¾ÑˆĞ¸Ğ±ĞºĞ°)
        plt.subplot(4, num_samples, i + 1 + 3 * num_samples)
        diff = np.abs(original[i] - denoised_display)
        plt.imshow(diff, cmap='hot')
        plt.title("Error")
        plt.axis('off')
        plt.colorbar()

    plt.suptitle(f"Ğ ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚Ñ‹ ÑˆÑƒĞ¼Ğ¾Ğ¿Ğ¾Ğ´Ğ°Ğ²Ğ»ĞµĞ½Ğ¸Ñ (Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¾ Ğ½Ğ° GPU)\n"
                f"MSE: {metrics['mse']:.6f}, "
                f"PSNR: {metrics['psnr']:.2f} dB, "
                f"SSIM: {metrics['ssim']:.4f}",
                fontsize=16)
    plt.tight_layout()
    plt.show()

def plot_training_history(history):
    """Ğ’Ğ¸Ğ·ÑƒĞ°Ğ»Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ Ğ¿Ñ€Ğ¾Ñ†ĞµÑÑĞ° Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ñ"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    # Loss
    ax1.plot(history.history['loss'], label='Training Loss')
    ax1.plot(history.history['val_loss'], label='Validation Loss')
    ax1.set_title('Model Loss')
    ax1.set_ylabel('Loss')
    ax1.set_xlabel('Epoch')
    ax1.legend()

    # MAE
    ax2.plot(history.history['mae'], label='Training MAE')
    ax2.plot(history.history['val_mae'], label='Validation MAE')
    ax2.set_title('Model MAE')
    ax2.set_ylabel('MAE')
    ax2.set_xlabel('Epoch')
    ax2.legend()

    plt.tight_layout()
    plt.show()

# 8. ĞÑ†ĞµĞ½ĞºĞ° Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸ Ğ½Ğ° Ñ‚ĞµÑÑ‚Ğ¾Ğ²Ñ‹Ñ… Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…
print("\nĞÑ†ĞµĞ½ĞºĞ° Ğ½Ğ° Ñ‚ĞµÑÑ‚Ğ¾Ğ²Ñ‹Ñ… Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…...")
denoised_test, test_mse, test_psnr, test_ssim = evaluate_model(
    autoencoder, x_test, x_test_noisy
)

test_metrics = {
    'mse': test_mse,
    'psnr': test_psnr,
    'ssim': test_ssim
}

# 9. Ğ’Ğ¸Ğ·ÑƒĞ°Ğ»Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ Ñ€ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚Ğ¾Ğ²
print("\nĞ’Ğ¸Ğ·ÑƒĞ°Ğ»Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ Ñ€ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚Ğ¾Ğ²...")
plot_training_history(history)
plot_comparison(x_test[:5], x_test_noisy[:5], denoised_test[:5], test_metrics)

# 10. Ğ¡Ñ€Ğ°Ğ²Ğ½ĞµĞ½Ğ¸Ğµ Ñ Ğ·Ğ°ÑˆÑƒĞ¼Ğ»ĞµĞ½Ğ½Ñ‹Ğ¼Ğ¸ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸ÑĞ¼Ğ¸
print("\nĞ¡Ñ€Ğ°Ğ²Ğ½ĞµĞ½Ğ¸Ğµ Ñ Ğ·Ğ°ÑˆÑƒĞ¼Ğ»ĞµĞ½Ğ½Ñ‹Ğ¼Ğ¸ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸ÑĞ¼Ğ¸ (Ğ±ĞµĞ· Ğ¾Ğ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ¸):")
noisy_mse = np.mean((x_test - x_test_noisy) ** 2)
noisy_psnr, noisy_ssim = calculate_metrics(x_test, x_test_noisy)

print(f"Ğ‘ĞµĞ· Ğ¾Ğ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ¸ - MSE: {noisy_mse:.6f}, PSNR: {noisy_psnr:.2f} dB, SSIM: {noisy_ssim:.4f}")
print(f"ĞŸĞ¾ÑĞ»Ğµ Ğ°Ğ²Ñ‚Ğ¾ÑĞ½ĞºĞ¾Ğ´ĞµÑ€Ğ° - MSE: {test_mse:.6f}, PSNR: {test_psnr:.2f} dB, SSIM: {test_ssim:.4f}")

# Ğ’Ñ‹Ñ‡Ğ¸ÑĞ»ĞµĞ½Ğ¸Ğµ ÑƒĞ»ÑƒÑ‡ÑˆĞµĞ½Ğ¸Ñ
mse_improvement = (noisy_mse - test_mse) / noisy_mse * 100
psnr_improvement = (test_psnr - noisy_psnr) / noisy_psnr * 100

print(f"\nĞ£Ğ›Ğ£Ğ§Ğ¨Ğ•ĞĞ˜Ğ• ĞšĞĞ§Ğ•Ğ¡Ğ¢Ğ’Ğ:")
print(f"MSE ÑƒĞ»ÑƒÑ‡ÑˆĞµĞ½Ğ° Ğ½Ğ°: {mse_improvement:.1f}%")
print(f"PSNR ÑƒĞ»ÑƒÑ‡ÑˆĞµĞ½Ğ° Ğ½Ğ°: {psnr_improvement:.1f}%")

# 11. Ğ”ĞµĞ¼Ğ¾Ğ½ÑÑ‚Ñ€Ğ°Ñ†Ğ¸Ñ Ğ½Ğ° Ğ¾Ñ‚Ğ´ĞµĞ»ÑŒĞ½Ñ‹Ñ… Ğ¿Ñ€Ğ¸Ğ¼ĞµÑ€Ğ°Ñ…
def show_detailed_examples(original, noisy, denoised, indices=[0, 1]):
    """ĞŸĞ¾ĞºĞ°Ğ·Ñ‹Ğ²Ğ°ĞµÑ‚ Ğ´ĞµÑ‚Ğ°Ğ»ÑŒĞ½Ñ‹Ğµ Ğ¿Ñ€Ğ¸Ğ¼ĞµÑ€Ñ‹ Ñ Ğ¼ĞµÑ‚Ñ€Ğ¸ĞºĞ°Ğ¼Ğ¸ Ğ´Ğ»Ñ ĞºĞ°Ğ¶Ğ´Ğ¾Ğ³Ğ¾ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ñ"""
    for idx in indices:
        orig_img = original[idx]
        noisy_img = noisy[idx]
        denoised_img = denoised[idx]

        # Ğ’Ñ‹Ñ‡Ğ¸ÑĞ»ÑĞµĞ¼ Ğ¼ĞµÑ‚Ñ€Ğ¸ĞºĞ¸ Ğ´Ğ»Ñ ÑÑ‚Ğ¾Ğ³Ğ¾ ĞºĞ¾Ğ½ĞºÑ€ĞµÑ‚Ğ½Ğ¾Ğ³Ğ¾ Ğ¸Ğ·Ğ¾Ğ±Ñ€Ğ°Ğ¶ĞµĞ½Ğ¸Ñ
        img_mse = np.mean((orig_img - denoised_img) ** 2)
        img_psnr = psnr(orig_img, denoised_img, data_range=1.0)
        img_ssim = ssim(orig_img, denoised_img, data_range=1.0, channel_axis=-1)

        fig, axes = plt.subplots(1, 3, figsize=(15, 5))

        # ĞÑ€Ğ¸Ğ³Ğ¸Ğ½Ğ°Ğ»
        axes[0].imshow(np.clip(orig_img, 0, 1))
        axes[0].set_title(f'Original Image {idx}')
        axes[0].axis('off')

        # Ğ—Ğ°ÑˆÑƒĞ¼Ğ»ĞµĞ½Ğ½Ğ¾Ğµ
        axes[1].imshow(np.clip(noisy_img, 0, 1))
        axes[1].set_title(f'Noisy Image {idx}')
        axes[1].axis('off')

        # Ğ’Ğ¾ÑÑÑ‚Ğ°Ğ½Ğ¾Ğ²Ğ»ĞµĞ½Ğ½Ğ¾Ğµ
        axes[2].imshow(np.clip(denoised_img, 0, 1))
        axes[2].set_title(f'Denoised Image {idx}\n'
                         f'MSE: {img_mse:.4f}, PSNR: {img_psnr:.2f}, SSIM: {img_ssim:.4f}')
        axes[2].axis('off')

        plt.tight_layout()
        plt.show()

print("\nĞ”ĞµÑ‚Ğ°Ğ»ÑŒĞ½Ñ‹Ğµ Ğ¿Ñ€Ğ¸Ğ¼ĞµÑ€Ñ‹:")
show_detailed_examples(x_test, x_test_noisy, denoised_test)

print("\nâ­ ĞĞ’Ğ¢ĞĞ­ĞĞšĞĞ”Ğ•Ğ  Ğ£Ğ¡ĞŸĞ•Ğ¨ĞĞ ĞĞ‘Ğ£Ğ§Ğ•Ğ ĞĞ GPU! â­")
```
